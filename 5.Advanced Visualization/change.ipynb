```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import warnings 
```

```python
#pandas 
movies = pd.read_csv('Movie-Ratings.csv')
movies.columns=['Film','Genre','CriticRatings','AudienceRatings','BudgetMillion','Year']
#
%matplotlib inline
#warnings
warnings.filterwarnings('ignore')
```

```python
movies.Genre = movies.Genre.astype('category')
movies.Genre.cat.categories
```

```python
list1  = list()

for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre == gen ].BudgetMillion)
```

```python
#1
list2  = list()
newlist2 = list()
for gen in movies.Genre.cat.categories:
    list2.append(movies[movies.Genre == gen ].BudgetMillion)
    newlist2.append(gen)
```

```python
sns.set_style('whitegrid')
fig, ax = plt.subplots()
fig.set_size_inches(11.7,8.27)  
h = plt.hist(list2, bins=25, stacked=True, label=newlist2)
plt.title('Movie Budget Distribution',fontsize=20,color='Green',fontname='Console')
plt.ylabel('Numbers of Movie',fontsize=20,color='Black')
plt.xlabel('Budget',fontsize=20,color='Aqua')
plt.legend()
plt.show()
```
