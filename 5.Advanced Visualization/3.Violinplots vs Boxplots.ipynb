```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import warnings 
```

```python
#pandas 
movies = pd.read_csv('Movie-Ratings.csv')
movies.columns=['Film','Genre','CriticRatings','AudienceRatings','BudgetMillion','Year']
#
%matplotlib inline
#warnings
warnings.filterwarnings('ignore')
```

```python
z = sns.violinplot(data=movies, x= 'Genre',y='CriticRatings') 
```

```python
w = sns.boxplot(data=movies, x= 'Genre',y='CriticRatings') 
```

```python
g = sns.violinplot(data=movies[movies.Genre=='Drama'], x= 'Year',y='CriticRatings') 
```

```python
h = sns.boxplot(data=movies[movies.Genre=='Drama'], x= 'Year',y='CriticRatings') 
```
