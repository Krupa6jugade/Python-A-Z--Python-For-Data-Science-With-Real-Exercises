```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import warnings 
```

```python
#pandas 
movies = pd.read_csv('Movie-Ratings.csv')
movies.columns=['Film','Genre','CriticRatings','AudienceRatings','BudgetMillion','Year']
#
%matplotlib inline
#warnings
warnings.filterwarnings('ignore')
```

JointPlots

```python
movies.Genre = movies.Genre.astype('category')
movies.Genre.cat.categories
```

```python
for gen in movies.Genre.cat.categories:
    print(gen)
```

```python
list1  = list()

for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre == gen ].BudgetMillion)
```

```python
print(list1)
```

```python
list2  = list()
newlist2 = list()
for gen in movies.Genre.cat.categories:
    list2.append(movies[movies.Genre == gen ].BudgetMillion)
    newlist2.append(gen)
    
h = plt.hist(list2, bins=25, stacked=True, label=newlist2)
plt.legend()
plt.show()
```
