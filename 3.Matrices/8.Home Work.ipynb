```python
def myplot(data,playerlist=Players):
    Col = {"KobeBryant": "Black", "JoeJohnson": "Red", "LeBronJames": "Green", "CarmeloAnthony": "Blue", "DwightHoward": "Pink", "ChrisBosh":"Aqua","ChrisPaul": "Purple", "KevinDurant": "Yellow", "DerrickRose": "Gray", "DwayneWade": "cyan"}
    Mar = {"KobeBryant": ",", "JoeJohnson": ".", "LeBronJames": "<", "CarmeloAnthony": ">", "DwightHoward": "^", "ChrisBosh":"H","ChrisPaul": "+", "KevinDurant": "x", "DerrickRose": "D", "DwayneWade": "o"}
    for name in playerlist:
        plt.plot(data[Pdict[name]],c =Col[name],ls='--',marker=Mar[name], ms=7,label=name)
    plt.legend(loc='upper left' , bbox_to_anchor = (1,1))
    plt.xticks(list(range(0,10)),Seasons,rotation='vertical')
    plt.show()
```
